```python
import requests
import pandas as pd
import time
```

```python
APP_ID = "90fb93ab"
APP_KEY = "ac0d74800544bfc63431a2d3747a92c5"
```

```python
def fetch_page(country, job_title, page):
    job_encoded = job_title.replace(" ", "%20")
    url = f"http://api.adzuna.com/v1/api/jobs/{country}/search/{page}?app_id={APP_ID}&app_key={APP_KEY}&results_per_page=50&what={job_encoded}&max_days_old=30&content-type=application/json"

    response = requests.get(url)
    data = response.json()

    if "results" in data:
        return data["results"]
    else:
        return []
```

```python
def fetch_job_data(country, job_title, pages=10):
    all_results = []

    for page in range(1, pages + 1):
        print(f"Fetching {job_title} | {country} | Page {page}")
        results = fetch_page(country, job_title, page)

        if not results:
            print("No more results.")
            break

        all_results.extend(results)
        time.sleep(1)  # avoid rate limits

    return pd.DataFrame(all_results)
```

```python
countries = ["gb", "de", "nl"]   # UK, Germany, Netherlands
job_titles = ["data analyst", "software engineer", "business analyst"]
```

```python
all_datasets = {}

for country in countries:
    for job in job_titles:
        print(f"\n=== Fetching {job.upper()} in {country.upper()} ===")
        df = fetch_job_data(country, job, pages=10)  # 10 pages = 500 jobs

        filename = f"{country}_{job.replace(' ', '_')}.csv"
        df.to_csv(filename, index=False)

        all_datasets[f"{country}_{job}"] = df

        print(f"Saved: {filename} ({len(df)} rows)")
```

# Uploading and Cleaning
GB - Data analyst


```python
df = pd.read_csv("/content/gb_data_analyst.csv")
df.head()
```

```python
df["description_clean"] = (
    df["description"]
    .astype(str)
    .str.replace("\n", " ", regex=False)
    .str.replace("\r", " ", regex=False)
    .str.replace("<.*?>", "", regex=True)
)
```

```python
import re

skills = [
    "python", "excel", "sql", "power bi", "tableau", "machine learning",
    "ai", "communication", "leadership", "cloud", "azure", "aws",
    "data analysis", "statistics", "project management", "java"
]

def extract_skills(text):
    found = []
    text_lower = text.lower()
    for skill in skills:
        if skill in text_lower:
            found.append(skill)
    return found

df["skills_found"] = df["description_clean"].apply(extract_skills)
```

```python
from collections import Counter

all_skills = Counter()

for row in df["skills_found"]:
    all_skills.update(row)

all_skills.most_common(20)
```
