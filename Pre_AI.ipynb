```python
import pandas as pd

eu = pd.read_csv("/content/sample_data/pre_ai/emed_careers_eu.csv")


eu_clean = pd.DataFrame()
eu_clean["job_title"] = eu["job_title"]
eu_clean["job_description"] = eu["job_description"]

# Extract country
eu_clean["country"] = eu["location"].str.extract(r"(UK|United Kingdom|Ireland|Europe)", expand=False)

# Convert date
eu_clean["date_posted"] = pd.to_datetime(eu["post_date"], errors="coerce")

# Add source
eu_clean["source"] = "emed_careers_eu"

# Role classifier
def classify_role(title):
    t = str(title).lower()
    if "accountant" in t:
        return "accountant"
    if "business" in t:
        return "business_analyst"
    if "data" in t:
        return "data_analyst"
    if "software" in t or "developer" in t or "engineer" in t:
        return "software_engineer"
    return None

eu_clean["role"] = eu_clean["job_title"].apply(classify_role)

# Drop missing essentials
eu_clean = eu_clean.dropna(subset=["job_title", "job_description", "country", "role"])

eu_clean.head()
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# Load raw NDJSON
df_raw = pd.read_json("/content/sample_data/pre_ai/techmap-jobs-export-2020-10_ie.json", lines=True)

# Flatten nested JSON
df = pd.json_normalize(df_raw.to_dict(orient="records"))

# Job title
df["job_title"] = df["json.schemaOrg.title"].fillna(df["position.name"])

# Job description
df["job_description"] = df["text"].apply(
    lambda x: BeautifulSoup(x, "html.parser").get_text() if isinstance(x, str) else None
)

# Country
df["country"] = df["sourceCC"]

# Date posted
df["date_posted"] = pd.to_datetime(df["dateCreated.$date.$numberLong"], unit="ms", errors="coerce")

# Source
df["source_clean"] = df["source"]

# Role classifier
def classify_role(title):
    t = str(title).lower()
    if "accountant" in t:
        return "accountant"
    if "business" in t:
        return "business_analyst"
    if "data" in t:
        return "data_analyst"
    if "software" in t or "developer" in t or "engineer" in t:
        return "software_engineer"
    return None

df["role"] = df["job_title"].apply(classify_role)

# Final cleaned dataset
tm2020_clean = df[[
    "job_title",
    "job_description",
    "country",
    "date_posted",
    "source_clean",
    "role"
]].dropna()

tm2020_clean.head()
```

```python
df_raw = pd.read_json("/content/sample_data/pre_ai/techmap-jobs-export-2021-10_ie.json", lines=True)
df = pd.json_normalize(df_raw.to_dict(orient="records"))

df["job_title"] = df["json.schemaOrg.title"].fillna(df["position.name"])

df["job_description"] = df["text"].apply(
    lambda x: BeautifulSoup(x, "html.parser").get_text() if isinstance(x, str) else None
)

df["country"] = df["sourceCC"]
df["date_posted"] = pd.to_datetime(df["dateCreated.$date.$numberLong"], unit="ms", errors="coerce")
df["source_clean"] = df["source"]
df["role"] = df["job_title"].apply(classify_role)

tm2021_clean = df[[
    "job_title",
    "job_description",
    "country",
    "date_posted",
    "source_clean",
    "role"
]].dropna()
```

```python
skill_dict = {
    "programming": [
        "python", "r", "java", "c++", "c#", "javascript", "matlab"
    ],
    "data_tools": [
        "sql", "excel", "tableau", "power bi", "sas", "stata", "spark"
    ],
    "cloud": [
        "aws", "azure", "gcp", "google cloud", "cloud computing"
    ],
    "ml_ai": [
        "machine learning", "deep learning", "nlp", "neural network"
    ],
    "software_dev": [
        "agile", "scrum", "devops", "git", "docker", "kubernetes"
    ],
    "soft_skills": [
        "communication", "teamwork", "leadership", "problem solving",
        "analytical", "stakeholder", "collaboration"
    ]
}
```

```python
import re

def extract_skills(text, skill_dict):
    text = str(text).lower()
    found_skills = []

    for category, keywords in skill_dict.items():
        for kw in keywords:
            if re.search(rf"\b{kw}\b", text):
                found_skills.append(kw)

    return list(set(found_skills))
```

```python
pre_ai_full["skills"] = pre_ai_full["job_description"].apply(
    lambda x: extract_skills(x, skill_dict)
)
```

```python
skills_exploded = pre_ai_full.explode("skills")
skills_exploded = skills_exploded.dropna(subset=["skills"])
```
